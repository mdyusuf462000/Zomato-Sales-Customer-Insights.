1 Step Data Claeing 

```python
#Columns name ko clean karna
df.columns = df.columns.str.strip().str.replace(" ", "_").str.replace("(for_two_people)", ("for two people"), regex=False)
df.columns
```

```python
df.head()
```

```python
#rate column ko float me change krna
df['rate'] = df['rate'].str.replace("/5","", regex=False).str.strip()
df['rate'] = pd.to_numeric(df['rate'],errors='coerce')
df['rate'].dtype
```

```python
df.head()
```

```python
df['approx_costfor two people'] = df['approx_costfor two people'].astype(str).str.replace(",", "").astype(float).astype(int)
df['approx_costfor two people']
```

```python
df['approx_costfor two people'].dtype
```

```python
# Top 10 Highest Restaurants

query = "SELECT DISTINCT name, rate, votes FROM zomato ORDER BY rate DESC, votes DESC LIMIT 10;"
df_top_rated = pd.read_sql(query, con=engine)
df_top_rated
```

```python
query = "SELECT online_order, COUNT(*) as count FROM zomato GROUP BY online_order;"
online_order_count = pd.read_sql(query, con=engine)
online_order_count
```

```python
#Low Price And best rateing Restaurants
query = "SELECT name, rate, `approx_costfor two people` FROM zomato WHERE `approx_costfor two people` < 500 AND rate > 4 ORDER BY rate DESC;"
df_affordable = pd.read_sql(query, con=engine)
df_affordable
```
